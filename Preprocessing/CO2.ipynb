# **HANDLING NULL VALUES**

```python
#Displaying only columns with null(missing) values with percentage
null_columns=df.columns[df.isnull().any()]
null_summary=df[null_columns].isnull().sum()
null_percentage = ((null_summary / len(df)) * 100).round(2)
null_dataframe=pd.DataFrame({'Column':null_columns,'Null Count':null_summary,'Percentage %':null_percentage}).sort_values(by='Percentage %',ascending=False)
display(null_dataframe)

```

IT SEEMS THERE IS A PATTERN IN NULL VALUES BETWEEN
*    MANURE APPLIED TO SOILS AND MANURE MANAGEMENT
*    FORESTLAND AND NET FOREST CONVERSION




```python
# Check correlation between the columns with potential patterns
correlation = df[['Manure applied to Soils', 'Manure Management']].corr()

# Display the correlation matrix
print(correlation)
```

```python
# Check correlation between the columns with potential patterns
correlation = df[['Forestland', 'Net Forest conversion']].corr()

# Display the correlation matrix
print(correlation)
```

THE ABOVE CORRELATION TEST SUGGESTS THAT THOSE TWO COLUMNS ARE CLOSELY RELATED AND MISSING VALUES ARE LIKELY ASSOCIATED WITH EACH OTHER.

FILLING THE NULL VALUES WITH 0

```python
for columns in null_columns:
  df.fillna({columns:0},inplace=True)
```

Let's consider the data after 2011(Occurence of Last boundary split). Because,
*   Older data might not reflect current trends, especially for countries that have undergone significant changes (e.g., splits, economic reforms).
*   Post-2011 data may align better with modern boundaries and reporting standards.
*   Many countries standardized their data reporting formats after 2010 due to international agreements like the Kyoto Protocol,Paaris agreement and subsequent updates.






```python
df_updated=df[df['year']>2011]
```

```python
df_updated['area'].unique()
```

```python
df_updated.loc[:, 'area'] = df_updated['area'].replace({
    'Congo': 'Republic of Congo',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Viet Nam': 'Vietnam',
    'Iran (Islamic Republic of)': 'Iran',
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Micronesia (Federated States of)': 'Micronesia',
    'Netherlands (Kingdom of the)': 'Netherlands',
    "Democratic People's Republic of Korea": 'North Korea',
    'Republic of Korea': 'South Korea',
    'Falkland Islands (Malvinas)': 'Malvinas'
})
```

Now let's Group Countries:

French Polynesia: Should be categorized under France (as it is an overseas territory).

Falkland Islands (Malvinas): This is a British overseas territory, and categorized under United Kingdom.

Pacific Islands Trust Territory: This refers to a former trust territory, and categorized under United States.

China, Hong Kong SAR
China, Macao SAR
China, mainland
China, Taiwan Province of : Categorised under China

```python
df_updated.loc[:,'area'] = df_updated['area'].replace({
    'China, Hong Kong SAR': 'China',
    'China, Macao SAR': 'China',
    'China, mainland': 'China',
    'China, Taiwan Province of': 'China',
    'French Polynesia': 'France',
    'Falkland Islands (Malvinas)': 'United Kingdom of Great Britain and Northern Ireland',
    'Pacific Islands Trust Territory': 'United States of America '
})
```

```python
df_updated=df_updated.reset_index(drop=True)
```
